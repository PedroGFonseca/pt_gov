```python
base_Url = 'http://servico.dados.gov.pt/v1/ama/AMinhaRuaOcorrencias'
end_Url = '?&format=json'
```

```python
import pandas as pd 
```

```python
def prepare_data(base_Url, end_Url):
    raw_data = pd.read_json(base_Url+end_Url);
    data = pd.DataFrame()
    for i in range(len(raw_data)): 
        series = pd.Series(raw_data.iloc[i][0])
        data[i] = series
    return data.T
```

```python
data = prepare_data(base_Url, end_Url)
```
